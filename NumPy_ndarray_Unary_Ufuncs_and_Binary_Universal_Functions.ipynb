## NumPy ndarray Along with Examples of Unary Ufuncs and Binary Universal Functions

Examples of Basic Functions using NumPy

```python
# import NumPy
import numpy as np
```

### Generate Nine Arrays via Random Number

```python
# three 3x3 arrays
A = np.random.randn(3,3)
B = np.random.randn(3,3)
C = np.random.randn(3,3)

# three 5x5 arrays
D = np.random.randn(5,5)
E = np.random.randn(5,5)
F = np.random.randn(5,5)

# three 7x7 arrays
G = np.random.randn(7,7)
H = np.random.randn(7,7)
I = np.random.randn(7,7)

# make a list called "array" that contains all nine arrays
# this is to enable looping
arrays = {"A": A, "B": B, "C": C, "D": D, "E": E, "F": F, "G": G, "H": H, "I": I}
```

```python
# print out all nine arrays
for key, array in arrays.items():
    print("\n" + key + ":")
    print(array)
```

### Table 4-3 : Unary ufuncs

##### 1. abs and sqrt

```python
# since we cannot find the square root of negative values, 
# we find their absolute values first before finding the square root
for key, array in arrays.items():
    print("\n" + key + ":")
    print(np.sqrt(np.abs(array)))
```

#### 2. modf and exp

```python
# modf will return the fractional and integral parts of arrays as separate arrays
# if we want to access the fractional part, we can use indexing of [0]
# in this section, I compute the exponent of each element based on their fractional part
for key, array in arrays.items():
    print("\n" + key + ":")
    print(np.exp(np.modf(array)[0]))
```

#### 3. sign

```python
# get the sign for each element in each array
# indicator:
# 1 --> positive
# 0 --> zero
# -1 --> negative

for key, array in arrays.items():
    print("\n" + key + ":")
    print(np.sign(array))
```

#### 4. ceil

```python
# get the smallest integer greater than or equal to the element in each array
for key, array in arrays.items():
    print("\n" + key + ":")
    print(np.ceil(array))
```

#### 5. sin 

```python
# get the sin value of each element in the arrays
# the np.sin function gives the output in radians
# I print the values in degrees, by converting the 
# radians to degree using np.pi / 180. (as shown in calculator)
for key, array in arrays.items():
    print("\n" + key + ":")
    print(np.sin(array * np.pi / 180.))
```

### Table 4-4: Binary Universal Functions

##### Since the functions are carried out element wise, the pair of arrays need to be of the same shape

#### 6. add and subtract 

```python
# example for 3x3 arrays
print("A + B:")
print(np.add(A,B))

# example for 5x5 arrays
print("\nD - E:")
print(np.subtract(D,E))

# example for 7x7 arrays
print("\nG + H - I")
print(np.subtract(np.add(G,H), I))
```

#### 7. multiply and divide

```python
# example for 3x3 arrays
print("A x C:")
print(np.multiply(A,C))

# example for 5x5 arrays
print("\nD / F:")
print(np.divide(D,F))

# example for 7x7 arrays
print("\nH x I / G")
print(np.divide(np.multiply(H,I), G))
```

#### 8. maximum and minimum

```python
# example for 3x3 arrays
print("Element-wise maximum of arrays B and C:")
print(np.maximum(B,C))

# example for 5x5 arrays
print("\nElement-wise maximum of arrays E and F:")
print(np.maximum(E,F))

# example for 7x7 arrays
print("\nElement-wise minimum of arrays G and I:")
print(np.minimum(G,I))
```

#### 9. copysign

```python
# copy sign of values in second argument to values in first argument

# example for 3x3 arrays
print("Copy sign of C to A:")
print(np.copysign(A,C))

# example for 5x5 arrays
print("\nCopy sign of F to D:")
print(np.copysign(D,F))

# example for 7x7 arrays
print("\nCopy sign of I to G:")
print(np.copysign(G,I))
```

#### 10. greater, less and equal

```python
# example for 3x3 arrays
print("Is C greater than B?")
print(np.greater(C,B))

# example for 5x5 arrays
print("\nIs F less than E?")
print(np.less(F,E))

# example for 7x7 arrays
print("\nIs H equal to I?")
print(np.equal(H,I))
```
